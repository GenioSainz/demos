```python
%load_ext autoreload
%autoreload 2

from data_models.test_SS_data     import get_data_test_SS
from data_models.test_SS_models   import get_models_test_SS

from pyampacity.utils.utils     import print_lists
import numpy  as np

# 'Example A'     MODE = 0 CIGRE601 Annex E1 E2
# 'Example B'     MODE = 1 CIGRE601 Annex E1 E2
# 'random  data'  MODE = 2
#  4.6.1-4.6.2    MODE = 3 IEEE738  Anenex 4.6.1 and 4.6.2 is almost equal as CIGRE Annex E1 E2 Example A
MODE = 0
```

```python
N_TEST = 10
if N_TEST==1: print_balance=True
else:         print_balance=False

print_balance=True
print_balance=False

list_I1, list_I2 = [], []
list_T1, list_T2 = [], []

for i in range(N_TEST):
    
    data = get_data_test_SS(MODE=MODE)
    model_cigre, model_ieee = get_models_test_SS(data)
    
    # input conductor temperature
    Tss = data['conductor_temperature']
    list_I1.append( model_cigre.SS_Thermal_Rating(print_out=print_balance)  )
    list_I2.append( model_ieee.SS_Thermal_Rating(print_out=print_balance)   )
    
    # input conductor current
    Iss = data['conductor_current']
    model_cigre.conditions.conductor_current = Iss
    model_ieee.conditions.conductor_current  = Iss
    
    list_T1.append( model_cigre.SS_Conductor_Temperature(print_out=print_balance)  )
    list_T2.append( model_ieee.SS_Conductor_Temperature(print_out=print_balance)   )
    
 
I12 = np.vstack((list_I1,list_I2))
dI  = np.diff(I12,axis=0).flatten().tolist() 
print('');print('SS-I (A)');print_lists(['cigre','ieeee','diff'],[list_I1,list_I2,dI])


T12 = np.vstack((list_T1,list_T2))
dT  = np.diff(T12,axis=0).flatten().tolist()
print('');print('SS-T (°C)');print_lists(['cigre','ieeee','diff'],[list_T1,list_T2,dT])

# data

```
