```python
%load_ext autoreload
%matplotlib widget
%autoreload 2

import pandas as pd
import numpy  as np
pd.set_option("display.max_rows", None)

df_imperial = pd.read_csv("data_ACSR_imperial.csv")
df_metric   = df_imperial.copy()
#df_imperial
```

```python
lb_1000ft_2_kg_km = (0.453592/0.0003048)/1000
inch2_2_mm2       = (2.54*10)**2
inch_2_mm         = 2.54*10
ohms_mi_2_ohms_km = 1/1.60934

df_metric['S_AL'] *= inch2_2_mm2
df_metric['S_T']  *= inch2_2_mm2

df_metric['d_AL'] *= inch_2_mm
df_metric['d_ST'] *= inch_2_mm
df_metric['D']    *= inch_2_mm
df_metric['D_c']  *= inch_2_mm

df_metric['W_T' ] *= lb_1000ft_2_kg_km 
df_metric['W_AL'] *= lb_1000ft_2_kg_km 
df_metric['W_ST'] *= lb_1000ft_2_kg_km 

df_metric['R25'] *= ohms_mi_2_ohms_km
df_metric['R50'] *= ohms_mi_2_ohms_km 
df_metric['R75'] *= ohms_mi_2_ohms_km 

# ROUND THE DATA
round_dict = {}
for i in [1,2,4,6,7,8,9,10,11]: round_dict[df_metric.columns[i]] = 2
for i in [12,13,14]:            round_dict[df_metric.columns[i]] = 5
df_metric = df_metric.round(round_dict)

df_metric
```
